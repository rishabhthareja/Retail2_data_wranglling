# Count the words in 'Description' column and add another column representing the numbers.

```python
num_Words=[]
for i in df['Description']:
    if type(i) == str:
        words = i.split()
        len_words = len(words)
        num_Words.append(len_words)
    
    else:
        num_Words.append('non string')
        
        
df["word_count of_descr"]=num_Words
        
df.head(5)
```

```python
df_new=df
df_new.dropna()
df_new['num_words1'] = [len(i) if type(i)==list else "nan" for i in df_new['Description'].str.split() ]

df_new['Description_1'] = [i for i in df_new['Description'].str.split()]

df_new
```
